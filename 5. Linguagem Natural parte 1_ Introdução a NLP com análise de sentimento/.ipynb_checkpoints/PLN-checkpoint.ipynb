# 2. Bag of Words: criando representações da linguagem humana.

![](https://i.imgur.com/zc3OhYl.png)

[Documentação](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) de CountVectorizer. 

Texto sobre [sparse matrix](https://pandas.pydata.org/pandas-docs/stable/user_guide/sparse.html#sparse-migration).

![](https://cmdlinetips.com/wp-content/uploads/2018/03/Sparse_Matrix.png)

```python
from sklearn.feature_extraction.text import CountVectorizer

texto = ["Assisti um filme ótimo", "Assisti um filme ruim"]  # Criando um corpus

# Vetorizando:
vetorizar = CountVectorizer(lowercase=False)
bag_of_words = vetorizar.fit_transform(texto)
bag_of_words
```

```python
vetorizar.get_feature_names_out()  

# Ele transforma tudo para letras minúsculas, logo vamos passar o parâmetro lowercase para o CountVectorizer
```

```python
matriz_esparsa = pd.DataFrame.sparse.from_spmatrix(bag_of_words, columns=vetorizar.get_feature_names_out())
matriz_esparsa
```

```python
vetorizar = CountVectorizer(lowercase=False)
bag_of_words = vetorizar.fit_transform(resenha.text_pt)
bag_of_words.shape
```

```python
# Segregando em dados de treino e teste

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
def classificar_texto(texto, coluna_texto, coluna_classificacao):
    vetorizar = CountVectorizer(lowercase=False, max_features=50)
    bag_of_words = vetorizar.fit_transform(texto[coluna_texto])

    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words, texto[coluna_classificacao], random_state=42)

    regressao_logistica = LogisticRegression()
    regressao_logistica.fit(treino, classe_treino)
    return regressao_logistica.score(teste, classe_teste)
```

```python
classificar_texto(resenha, "text_pt", "classificacao")
```

# 3. Visualizando os dados com WordCloud

DOCUMENTAÇÃO: [WordCloud](https://github.com/amueller/word_cloud)

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

todas_palavras = ' '.join([texto for texto in resenha.text_pt])
len(todas_palavras)
```

```python
nuvem_palavras = WordCloud(width=800, height=500, 
                           max_font_size = 110, 
                           collocations = False).generate(todas_palavras)
```

```python
plt.figure(figsize=(10,7))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
def nuvem_palavras_neg(texto, coluna_texto):
    texto_negativo = texto.query("sentiment == 'neg'")
    todas_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])

    nuvem_palavras = WordCloud(width= 800, height= 500,
                              max_font_size = 110,
                              collocations = False).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
def nuvem_palavras_pos(texto, coluna_texto):
    texto_positivo = texto.query("sentiment == 'pos'")
    todas_palavras = ' '.join([texto for texto in texto_positivo[coluna_texto]])

    nuvem_palavras = WordCloud(width= 800, height= 500,
                              max_font_size = 110,
                              collocations = False).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
nuvem_palavras_neg(resenha, "text_pt")
```

```python
nuvem_palavras_pos(resenha, "text_pt")
```
